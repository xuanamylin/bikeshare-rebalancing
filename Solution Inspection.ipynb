```python
# Function to load data
def load_pickle(file_path):
    with open(file_path, 'rb') as handle:
        content = pickle.load(handle)
    return content

# Function to print solutions
def print_solution(file_path):
    solution = load_pickle(output_path + file_name)
    
    # Objective
    print("Unsatisfied Demand (to minimize)")
    print("SA: {}".format(solution['sa']['unsatisfied_customers']))
    print("ACO: {}".format(solution['aco']['unsatisfied_customers']))
    
    # Time
    print("Time")
    print("SA: {}".format(solution['sa']['time']))
    print("ACO: {}".format(solution['aco']['time']))

    print("\nSA Route")
    for i, df in solution['sa']['route_df'].items():
        print("Truck {}:".format(i))
        display(df)
        
    print("ACO Route")
    for i, df in solution['aco']['route_df'].items():
        print("Truck {}:".format(i))
        display(df)

    print("\nSA Station Inventory")
    display(solution['sa']['station_inv_df'])
    print("ACO Station Inventory")
    display(solution['aco']['station_inv_df'])
```

**Notes:**
* The cases come from historical data. The solution files are named using the template `solution_<num>stops_<case_date>_<case_time_window>_<versin>`.
* 'redist' column: the number of bikes at station after redistribution
* 'expected' column: the number of bikes expected at station based on forecasted inflows and outflows
* 'actual' column: the number of bikes current at the station; extracted from sensor data
* In SA, action < 0 means pick-ups; In ACO, action > 0 means pickups. Ouch...
